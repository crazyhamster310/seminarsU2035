Создадим массив косиносуов и сикнусов от -4pi до 4pi. Визаулизируем его и прповерим основное тригонометрическое тождество.

```python
a = np.linspace(-4*np.pi, 4*np.pi, 100)
```

```python
plt.plot(np.cos(a), label="cos")
plt.plot(np.sin(a), label="sin")
plt.legend()
```

```python
np.any((np.cos(a)**2 + np.sin(a)**2).round() == 1)
```

Посчитаем кумулятивную сумму и среднее

```python
a = np.array([1, 3, 4, 10])

np.cumsum(a) / np.arange(1, 5)
```
