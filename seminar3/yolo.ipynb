```python
from ultralytics import YOLO
import matplotlib.pyplot as plt
```

#### Вам нужно загрузить веса в YOLO и сделать предсказания для test_img.png. После этого выполните последнюю ячейку, чтобы отобразить результат

```python
model = YOLO('путь к файлу весов (.pt)')
```

```python
res = model.predict("test_img.png", conf=0.8)
```

```python
plt.imshow(res[0].plot())
```
